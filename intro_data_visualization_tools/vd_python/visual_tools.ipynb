```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Matplotlib - Basic Plots

### Figure and an Axes

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()
```

```python
"""
A Figure, uma instância da classe plt.Figure, pode ser pensada como
um único contêiner que contém todos os objetos que representam eixos,
gráficos, texto e rótulos.
"""

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
"""
Para criar uma única figura com várias linhas é necessário
apenas chamar a função plot várias vezes.
"""
# x = np.linspace(0, 20, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
"""
Para ajustar a cor basta inserir o parâmetro 'color', que aceita
uma string como argumento e que representa qualquer cor imaginável. 
A cor pode ser especificada de várias maneiras, como segue:
"""

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.figure(figsize=(15,10))
x = np.linspace(0, 10, 1000)
plt.subplot(2, 2, 1)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# plt.plot(x, x + 4, linestyle='-')  # solid
# plt.plot(x, x + 5, linestyle='--') # dashed
# plt.plot(x, x + 6, linestyle='-.') # dashdot
# plt.plot(x, x + 7, linestyle=':');  # dotted
  
plt.subplot(2, 2, 2)
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r')  # dotted red
  
plt.show()
```

###  Axes Limits

```python
"""
A maneira mais básica de ajustar os limites do eixo
é usar plt.xlim e plt.ylim 
"""
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.figure(figsize=(15,10))
x = np.linspace(0, 10, 1000)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))

# Para exibir um eixo ao contrário, 
# basta inverter a ordem dos argumentos:
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);

plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x))

# def. os limites de x e y em uma única chamada 
# passando [xmin, xmax, ymin, ymax] como parâmetro
plt.axis([-1, 11, -1.5, 1.5]);

plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x))

# estreitar automaticamente os limites em torno do gráfico atual
plt.axis('tight');
  
plt.subplot(2, 2, 4)
plt.plot(x, np.sin(x))

# garantir a proporção na tela de que uma unidade em x 
# seja igual a uma unidade em y
plt.axis('equal');


plt.show()
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))

plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
"""
Quando várias linhas estão sendo mostradas em um único eixo, 
é bastante útil criar uma legenda de plotagem como rótulo p/ cada tipo de linha.
Há um built-in method que cria rapidamente esses rótulos.
"""

plt.figure(figsize=(15,10))
x = np.linspace(0, 10, 1000)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();

plt.subplot(2, 2, 2)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True, loc='upper left');


plt.show()
```

## Scatter Plots

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black');

plt.subplot(2, 2, 2)
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);


plt.show()
```

```python
"""
O terceiro argumento na chamada de função é um caractere que representa o tipo
de símbolo usado para a plotagem. Então, como é possível controlar o estilo da
linha atráves de códigos de string curtos, também é possível controlar o estilo 
do marcador. A seguir as possibilidades de marcadores mais comuns: 
"""

plt.figure(figsize=(10,8))
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1);
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.scatter(x, y, marker='o');

plt.subplot(2, 2, 2)
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale


plt.subplot(2, 2, 3)
plt.scatter(x, y, marker='>', color='purple');
  
plt.subplot(2, 2, 4)
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='inferno')
plt.colorbar();  # show color scale


plt.show()
```

## Histograms

```python
plt.style.use('seaborn-white')

data = np.random.randn(1000)
plt.hist(data);

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Example of Histogram Plot")
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='black');

plt.subplot(2, 2, 2)
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

# kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40, edgecolor='black')

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)

plt.show()
```

## Subplots

```python
# Simple Grids of Subplots

"""
O método subplot aceita três argumentos, sendo eles: o número de linhas, o número de colunas
e o índice do gráfico a ser criado neste esquema, que vai do canto superior esquerdo
ao canto inferior direito. Ou seja, plt.subplot(rows, columns, index)
"""

plt.figure(figsize=(12,6))

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

```python
"""
Utilizando o método subplots_adjust com os argumentos hspace e wspace 
para especificar o espaçamento ao longo da altura e largura da figura.
Neste caso, o espaço é 40% da largura e altura. 
"""

fig = plt.figure()
# fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')  
    
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

plt.figure(figsize=(12,6))

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

## Seaborn Quick Example

```python
import seaborn as sns
```

```python
sns.set_theme()

uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
```

```python
plt.figure(figsize=(13,13))
fig.subplots_adjust(hspace=1, wspace=0.8)

colormap = ['coolwarm', 'YlGnBu', 'inferno',  'PuBuGn', 'BuGn', 'YlGn']

for i in range(1, 7):
    plt.subplot(3, 2, i)
    uniform_data = np.random.rand(10, 20)
    
    ax = sns.heatmap(uniform_data, 
                     vmin=0, vmax=1, 
                     xticklabels=2, yticklabels=False,
                     cmap=colormap[i - 1])
    plt.xlabel(f'Heatmap {i}', fontsize = 14)
```
